### 14) Now re-assign the country column in the DataFrame to `USA` if the row is a state.
Also add the state name as a new column.

```python
country={}

for x in range(0,len(countries)):
  
    if str(stateornot[x]).find('notastate')<1:
        
        country[x]='USA'
        
    else:
        country[x]=countries[x]

earthquakes['country']=country.values()
earthquakes.head()
```
